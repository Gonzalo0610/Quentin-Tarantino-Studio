```python
import statsmodels.api as sm

X = cleanreg.drop(columns=["original_title"])
y = cleanreg["vote_average"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
from sklearn.linear_model import LinearRegression

X = cleanreg.drop(columns=["original_title"])
y = cleanreg["vote_average"]

reg_model = LinearRegression().fit(X, y)

coefficients_table = pd.DataFrame({'Feature': X.columns, 'Coefficient': reg_model.coef_})
print(coefficients_table)
```

```python
df = pd.DataFrame({'y': cleanreg['vote_average'], 
                   'x1': cleanreg['budget'], 
                   'x2': cleanreg['runtime'], 
                   'x3': cleanreg['death'],
                   'x4': cleanreg['Position']})
df = sm.add_constant(df)

model = sm.OLS(df['y'], df[['const', 'x1', 'x2', 'x3', 'x4']])

results = model.fit()

print(results.summary())
```
